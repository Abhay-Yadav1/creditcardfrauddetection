```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
print(df['Class'].value_counts())

# Plot fraud vs. legitimate amounts
plt.figure(figsize=(10, 6))
plt.scatter(df[df['Class']==0].index, df[df['Class']==0]['Amount'], 
            alpha=0.5, label='Legitimate', s=2)
plt.scatter(df[df['Class']==1].index, df[df['Class']==1]['Amount'], 
            color='red', label='Fraud', s=2)
plt.xlabel('Transaction Index')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```

```python
df.describe(include='all')
```

```python
from sklearn.preprocessing import RobustScaler
rb=RobustScaler()
df[['Time','Amount']]=rb.fit_transform(df[['Time','Amount']])
```

```python
import numpy as np

# 1. Time-based features (convert seconds to hours)
df['hour_of_day'] = (df['Time'] % (24*3600)) // 3600  # 0-23
```

```python
x=df.drop(['Class', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28','V20','hour_of_day','V19','V16','V15','V11'],axis = 1)
y=df['Class']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
xtrain_res,ytrain_res=smote.fit_resample(xtrain,ytrain)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(xtrain_res)
X_test_scaled = scaler.transform(xtest)  # Use same scaler!
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,classification_report
```

```python
rfc=RandomForestClassifier(
    class_weight={0:1,1:100},
    random_state=42,
    n_estimators=200,
    n_jobs=-1
)
rfc.fit(X_train_scaled,ytrain_res)
```

```python
y_pred=rfc.predict(X_test_scaled)
print(classification_report(y_pred,ytest))
accuracy_score(y_pred,ytest)
```

```python
models = {
    # Tree-Based
    "Decision Tree": DecisionTreeClassifier(class_weight={0:1, 1:10}, max_depth=5, random_state=42),
    "CatBoost": CatBoostClassifier(auto_class_weights='Balanced', iterations=200, depth=6, silent=True, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5, weights='distance', n_jobs=-1),
    "SVM": SVC(class_weight={0:1, 1:10}, kernel='rbf', C=1.0, gamma='scale', random_state=42),
    "Naive Bayes": GaussianNB(),
    "Neural Net": MLPClassifier(hidden_layer_sizes=(50,), alpha=0.01, early_stopping=True, random_state=42)
}
from sklearn.metrics import  precision_recall_curve, auc

results = []
for name, model in models.items():
    # Train
    model.fit(X_train_scaled, ytrain_res)
    
    # Predict
    y_pred3 = model.predict(X_test_scaled)
    
    # Metrics
    report = classification_report(ytest, y_pred3, output_dict=True)
    precision = report['1']['precision']
    recall = report['1']['recall']
    f1 = report['1']['f1-score']
    
    # Store results
    results.append({
        'Model': name,
        'Precision': round(precision, 3),
        'Recall': round(recall, 3),
        'F1-Score': round(f1, 3)
    })

# Show results
results_df = pd.DataFrame(results).sort_values('Recall', ascending=False)
print(results_df)
```
